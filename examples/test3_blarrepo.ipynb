````python
import json
from IPython.display import Markdown, display

def print_python(python_text, indent=0):
    """This function prints python text in ipynb nicely formatted."""
    indented_text = "\n".join(" " * indent + line for line in python_text.splitlines())
    display(Markdown("```python\n" + indented_text + "\n```"))

def pretty_print_with_code(data, indent=0):
    if isinstance(data, dict):
        print(" " * indent + "{")
        for key, value in data.items():
            if key == "text":
                print(" " * (indent + 2) + f'"{key}": """')
                print_python(value, indent + 6)
                print(" " * (indent + 2) + '"""')
            else:
                if isinstance(value, dict):
                    print(" " * (indent + 2) + f'"{key}": {{')
                    pretty_print_with_code(value, indent + 4)
                    print(" " * (indent + 2) + "}")
                else:
                    print(" " * (indent + 2) + f'"{key}": "{value}",')
        print(" " * indent + "}")

for idx in range(5, 7):
    pretty_print_with_code(nodes[idx])
````
